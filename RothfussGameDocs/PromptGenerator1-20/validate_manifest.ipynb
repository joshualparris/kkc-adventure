```python
from pathlib import Path
import generate_prompts as gen

ROOT = Path(r"c:\Users\joshu_w0zb8cp\Documents\Project-Hub\RothfussGame\RothfussGameDocs\PromptGenerator1-20")
manifest_path = ROOT / "prompt_manifest_41_100.json"
output_dir = ROOT / "generated_prompts"

template = gen.read_text(ROOT / "prompt_template.md")
manifest_data = gen.load_manifest(manifest_path)
entries = gen.select_entries(manifest_data, 61, 100, None)
global_context = gen.build_global_context()
output_dir.mkdir(parents=True, exist_ok=True)
for entry in entries:
    values = gen.build_values(entry, global_context)
    rendered = gen.render_template(template, values)
    filename = f"prompt_{entry['number']:02}_{values['filename_slug']}.md"
    path = output_dir / filename
    path.write_text(rendered, encoding="utf-8")
    print(f"Wrote {path}")

print(f"Generated {len(entries)} prompts")
```
